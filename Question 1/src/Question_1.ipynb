Importing libraries and loading the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('train.csv')
df.head()
```

a) Fix missing values

```python
# Checking missing values
df.isnull().sum()

# Filling numerical missing values with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Filling categorical missing values with mode
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
df.head()
```

What I did:

1. I checked all columns to see how many missing values were present.

2. I selected numeric columns and filled their missing values using the median.

3. I selected categorical columns and replaced missing values with the mode.

4. I verified that the dataset had no more missing or empty cells.

Interpretation:

1. This step prevents errors because models cannot work with missing data.

2. Using the median preserves the original pattern without being affected by outliers.

3. Filling categories with the most common value keeps the data consistent.

4. The dataset becomes complete, clean, and reliable for the next steps.
